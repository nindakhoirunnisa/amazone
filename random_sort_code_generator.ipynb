```python
from random import randint
import json

def random_with_N_digits(n):
    range_start = 10**(n-1)
    range_end = (10**n)-1
    return randint(range_start, range_end)

data = {}
first_names = ['Kim', 'Kim', 'Min', 'Jung', 'Park', 'Kim', 'Jeon', 'Kim', 'Kim', 'Park', 'Lalisa', 'Kim', 'Pham', 'Mo', 'Kang', 'Lee', 'Lee', 'Lee', 'Im', 'Park']
last_names = ['Namjoon', 'Seokjin', 'Yoongi', 'Hoseok', 'Jimin', 'Taehyung', 'Jungkook', 'Jisoo', 'Jennie', 'Rose', 'Manoban', 'Minji', 'Hanni', 'Jihye', 'Haerin', 'Hyein', 'Jieun', 'Jungsuk', 'Yoona', 'Joohyun']
emails = []
phones = []
genders = ["Male", "Male", "Male", "Male", "Male", "Male", "Male", "Female", "Female", "Female"]
ages = [29, 31, 30, 29, 28, 28, 26, 28, 27, 26]

for i in range(10):
    x = random_with_N_digits(3)
    y = random_with_N_digits(3)
    z = random_with_N_digits(3)
    phone = (str(x)+"-"+str(y)+"-"+str(z))
    email = last_names[i].lower()
    emails.append(email+"@gmail.com")
    phones.append(phone)
    data['first_name'] = first_names[i]
    data['last_name'] = last_names[i]
    data['gender'] = genders[i]
    data['age'] = ages[i]
    data['phone'] = phones[i]
    data['email'] = emails[i]
    print(json.dumps(data, indent=4, sort_keys=True))
```
